```python
import polars as pl
import os
```

```python
# Get single cell tool inputs for cells likely to be in each cell cycle stage

output_dir = "/mnt/c/Users/peree/OneDrive/Desktop/Compbio_Code/markerproject_redux/cell_cycle_classification/Dad2/labelled_cells"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

cell_nuclear_df = pl.read_csv("/home/alex/alex_files/markerproject_redux/cell_cycle_classification/Dad2/all_cell_and_nuclear_features_for_cell_cycle_classification.csv")

# Likely G1
likely_g1 = (
    cell_nuclear_df
    .filter(
        (pl.col("Cell_Children_Nuclei_Count") == 1) & 
        (pl.col("Cell_MajorAL_Over_Minor_AL") <= 1.2)
        )
    .select(["Cell_ID", "Image_Path", "Center_X", "Center_Y"])
    .unique()
    .sample(n=3000, with_replacement=False, shuffle=True, seed=1705)
    .write_csv(f"{output_dir}/likely_g1_cells.csv")
    )


# Likely S/G2
likely_sg2 = (
    cell_nuclear_df
    .filter(
        (pl.col("Cell_Children_Nuclei_Count") <= 2) & 
        (pl.col("Cell_MajorAL_Over_Minor_AL") >= 1.45)
        )
    .select(["Cell_ID", "Image_Path", "Center_X", "Center_Y"])
    .unique()
    .sample(n=3000, with_replacement=False, shuffle=True, seed=1705)
    .write_csv(f"{output_dir}/likely_sg2_cells.csv")
    )


# Likely M/A
likely_ma = (
    cell_nuclear_df
    .filter(
        (pl.col("Cell_Children_Nuclei_Count") <= 2) & 
        (pl.col("Cell_MajorAL_Over_Minor_AL") >= 1.45) &
        (pl.col("Cell_Mean_Nuclei_AreaShape_Eccentricity") >= 0.65)
        )
    .select(["Cell_ID", "Image_Path", "Center_X", "Center_Y"])
    .unique()
    .sample(n=3000, with_replacement=False, shuffle=True, seed=1705)
    .write_csv(f"{output_dir}/likely_ma_cells.csv")
    )


# Likely T
likely_t = (
    cell_nuclear_df
    .filter(
        (pl.col("Cell_Children_Nuclei_Count") == 2) & 
        (pl.col("Cell_MajorAL_Over_Minor_AL") >= 1.5) &
        (pl.col("Cell_Mean_Nuclei_AreaShape_Eccentricity") >= 0.7)
        )
    .select(["Cell_ID", "Image_Path", "Center_X", "Center_Y"])
    .unique()
    .sample(n=3000, with_replacement=False, shuffle=True, seed=1705)
    .write_csv(f"{output_dir}/likely_t_cells.csv")
    )
```
